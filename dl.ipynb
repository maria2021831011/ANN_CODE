Single Layer (fixed weights, no random)

X (2 features)
   │
   ▼
Single Neuron (Weighted Sum)
   │   z = w·x + b
   ▼
Sigmoid Activation
   │
   ▼
Output (Probability → 0 or 1)

```python
import numpy as np
import pandas as pd
df = pd.read_csv(r"D:\prac\diabetes (1).csv")
data = df.values

X = data[:, :2]
Y = data[:, -1]

X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

w = np.array([-1.0, -1.0])
b = 0.0
lr = 0.01

def sigmoid(z):
    return 1 / (1 + np.exp(-z))


for epoch in range(20):

    loss = 0
    correct = 0

    for i in range(len(X)):

        x = X[i]
        y = Y[i]

        
        z = np.dot(w, x) + b
        y_hat = sigmoid(z)

        
        error = y_hat - y
        loss += error ** 2

        
        pred = 1 if y_hat >= 0.5 else 0
        if pred == y:
            correct += 1

    
        grad = 2 * error * y_hat * (1 - y_hat)

        
        w -= lr * grad * x
        b -= lr * grad

    acc = (correct / len(X)) * 100

    print("Epoch:", epoch,
          "Loss:", loss / len(X),
          "Accuracy:", acc)
```

Hidden Layer ANN (random weights,BIAS

```python
import numpy as np
import pandas as pd

df = pd.read_csv(r"D:\prac\diabetes (1).csv")
data = df.values

X = data[:, :2]
Y = data[:, -1]


X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

np.random.seed(42)


W1 = np.random.randn(2, 2)
b1 = np.random.randn(2)


W2 = np.random.randn(2)
b2 = np.random.randn()

lr = 0.01


def sigmoid(z):
    return 1 / (1 + np.exp(-z))


for epoch in range(20):

    loss = 0
    correct = 0

    for i in range(len(X)):

        x = X[i]
        y = Y[i]

        z1 = np.dot(x, W1) + b1
        a1 = sigmoid(z1)

       
        z2 = np.dot(a1, W2) + b2
        y_hat = sigmoid(z2)

        
        error = y_hat - y
        loss += error ** 2

      
        dz2 = 2 * error * y_hat * (1 - y_hat)

       
        da1 = dz2 * W2
        dz1 = da1 * a1 * (1 - a1)

       

        W2 -= lr * dz2 * a1
        b2 -= lr * dz2

        W1 -= lr * np.outer(x, dz1)
        b1 -= lr * dz1

        # accuracy
        pred = 1 if y_hat >= 0.5 else 0
        if pred == y:
            correct += 1

    acc = (correct / len(X)) * 100

    print("Epoch:", epoch,
          "Loss:", loss / len(X),
          "Accuracy:", acc)
```

X (2 features)
   │
   ▼
Hidden Layer (2 neurons)
   │   W1 (2×2), b1 (2)
   ▼
Activation (Sigmoid)
   │
   ▼
Output Layer (1 neuron)
   │   W2 (2×1), b2
   ▼
Sigmoid → Prediction (0/1)
